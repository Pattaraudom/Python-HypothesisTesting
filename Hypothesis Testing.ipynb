### 1.	My partner and I have recently rediscovered a love for Lego™. Not the children’s toy, but the really complex Lego with thousands of pieces, Star Wars themes, and fancy architecture (OK – fine – it’s still a children’s toy). Sometimes, we build Lego together, other times I build on my own. I would not admit this to her (at least, not without sufficient evidence), but I am pretty sure I am faster at building Lego alone than we are together. I have been collecting data on the time it takes to build the set, the number of pieces, the difficulty of the set, the number of differently coloured pieces, and who built it. 

```python
#Import the required libraries
import pandas as pd
import numpy as np
import os.path as osp
import statsmodels.api as sm
from statsmodels.formula.api import ols
from sklearn.linear_model import LinearRegression

#Import Data from Excel file
data_path = osp.join(osp.curdir,'MFIT860_Exam_Data.xlsx')
Lego_data = pd.read_excel(data_path,sheet_name='Lego')
Lego_data.head()
```

```python
#We create Builder as dummy variable
Builder_dummy= pd.get_dummies(Lego_data['Builder']) 
Lego_data= pd.concat([Lego_data,Builder_dummy],axis=1)
Lego_data= Lego_data.drop('Builder',axis=1)
Lego_data.head()
```

#### a.	I believe that I am faster at building Lego than my partner and I are together in relation to the number of pieces.  Conduct the most powerful test you can to see if this is the case.  Briefly explain the test and your results (i.e. state the null, alternative, the p-value of the result and what it means.) 2 marks

```python
#Build our model 
Lego_model = ols('Time_S ~ Num_Pieces + Num_Colour + Difficulty + Me',Lego_data).fit()
Lego_model.summary()
```

```python
Lego_hypothesis1 = '(Num_Pieces=0, Me=0)'
print(Lego_model.wald_test(Lego_hypothesis1))
```

#### b.	Now conduct a Chow test to determine if us building together and me by myself are different at all.  Briefly explain these results, particularly in light of your test above. 3 marks

```python
#Create new column 
Lego_data['C1'] = Lego_data['Num_Pieces'] * Lego_data['Me']
Lego_data['C2'] = Lego_data['Num_Colour'] * Lego_data['Me']
Lego_data['C3'] = Lego_data['Difficulty'] * Lego_data['Me']
Lego_data.head()
```

```python
#Build the model
Lego_modelb = ols('Time_S ~ Num_Pieces + Num_Colour + Difficulty + Me + C1 + C2 + C3',Lego_data).fit()
Lego_modelb.summary()
```

```python
#Now run the chow test which is just a specific joint hypothesis test!
hypothesisb = '(Me = 0, C1 = 0, C2 = 0, C3 = 0)'
print(Lego_modelb.wald_test(hypothesisb))
```

### Answer
P-value is at 0.6602622377862143 which is greater than alpha at 0.05.  we cannot reject the null hypothesis. This indicate that there is no different between building together and me by myself

##### c.	Your client has provided a dataset with some missing data and has told you to just impute all the missing values with the mean. What would you tell them? 3 marks

### Answer
There are 3 patterns of missing data and each pattern can be solve with different techniques. No every pattern of missing data can solve by mean substitution. Therefore, we should firstly identify pattern of missing data as following.

Here are 3 paterns of missing data 
1. Missing completely at random (MCAR) :there is no pattern of how data is missing 
2. Missing at random (MAR):  there is a pattern of how data is missing whihc can be explained by the available data. 
3. Missing not at random: there is a pattern of how data is missing whihc can be explained by other missing data. 

There are at least 3 categories of techniques can be used to deal with missing data
1. Drop observation : this tecnique only work with MCAR.If the sample is large enough that the loss of the MCAR observations from the sample have a small impact, this is certainly the easiest way to go.
2. Mean Substitution : we can substute the missing data with mean of the field.This technique is acceptable for MCAR
3. Modelling techniques : we develop model in software to estimate and assign missing data based on the remaining data
